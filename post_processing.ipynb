# Toxicity Filtering

```python
import pandas as pd
from detoxify import Detoxify
from tqdm import tqdm
import torch
from collections import Counter
import matplotlib.pyplot as plt
import numpy as np
```

```python
def analyze_token_lengths(strings):
    """
    Analyze the distribution of token (word) lengths in a list of strings.
    Uses simple whitespace splitting for tokenization.
    
    Parameters:
    strings (list): List of strings to analyze
    
    Returns:
    tuple: (basic_stats, length_frequencies, total_tokens)
    """
    # Calculate token lengths for each string
    lengths = []
    total_tokens = 0
    model = Detoxify('original')
    for text in tqdm(strings):
        # Split on whitespace and filter out empty strings
        tokens = model.tokenizer.encode(text)
        num_tokens = len(tokens)
        lengths.append(num_tokens)
        total_tokens += num_tokens
    
    # Basic statistics
    stats = {
        'mean': np.mean(lengths),
        'median': np.median(lengths),
        'std': np.std(lengths),
        'min': min(lengths),
        'max': max(lengths),
        'total_strings': len(strings),
        'total_tokens': total_tokens,
        'avg_tokens_per_string': total_tokens / len(strings),
        "more_512": len([i for i in lengths if i > 512])
    }
    
    # Calculate frequency distribution
    freq_dist = Counter(lengths)
    
    # Convert to sorted list of tuples
    freq_dist = sorted(freq_dist.items())
    
    return stats, freq_dist

def plot_token_distribution(freq_dist):
    """
    Create a histogram of token counts.
    
    Parameters:
    freq_dist (list): List of (token_count, frequency) tuples
    """
    token_counts, freqs = zip(*freq_dist)
    
    plt.figure(figsize=(12, 6))
    plt.bar(token_counts, freqs)
    plt.xlabel('Number of Tokens (Words)')
    plt.ylabel('Frequency')
    plt.title('Distribution of Token Counts per String')
    plt.grid(True, alpha=0.3)
    
    # Add a rolling average line if there are enough data points
    if len(token_counts) > 5:
        window_size = 5
        rolling_avg = np.convolve(freqs, np.ones(window_size)/window_size, mode='valid')
        plt.plot(token_counts[window_size-1:], rolling_avg, 'r-', label='Rolling Average')
        plt.legend()
    
    plt.show()

def filter_toxic_dialogs(input_file, output_file, batch_size=64, device='cuda'):
    """
    Filter out toxic dialogs while preserving the original DataFrame structure.
    
    Args:
        input_file (str): Path to input jsonl file
        output_file (str): Path to output jsonl file
        batch_size (int): Number of items to process in each batch
        device (str): 'cuda' or 'cpu'
    """
    # Load the original DataFrame
    df = pd.read_json(input_file, lines=True)
    
    # Initialize model
    model = Detoxify('original', device=device)
    
    # Convert turns to strings for toxicity checking
    all_dialogs = []
    dialog_indices = []  # Keep track of original indices
    
    # Process each row's turns into a single string
    for idx, turns in enumerate(tqdm(df["turns"], desc="Processing dialogs")):
        concatenated_turns = []
        for entry in turns:
            if entry.get("COMPUTER") is not None:
                concatenated_turns.append(f"COMPUTER: {entry['COMPUTER']}")
            if entry.get("HUMAN") is not None:
                concatenated_turns.append(f"HUMAN: {entry['HUMAN']}")
        
        dialog = " ".join(concatenated_turns)
        all_dialogs.append(dialog)
        dialog_indices.append(idx)
    
    # Initialize list for keeping track of non-toxic indices
    safe_indices = []
    
    # Process in batches
    for i in tqdm(range(0, len(all_dialogs), batch_size), desc="Checking toxicity"):
        batch = all_dialogs[i:i + batch_size]
        batch_indices = dialog_indices[i:i + batch_size]
        
        # Get predictions for batch
        with torch.no_grad():
            batch_results = model.predict(batch)
        
        # Check each result in the batch
        for j, idx in enumerate(batch_indices):
            # Check if all toxicity scores are below threshold
            single_result = {
                'toxicity': batch_results['toxicity'][j],
                'severe_toxicity': batch_results['severe_toxicity'][j],
                'obscene': batch_results['obscene'][j],
                'threat': batch_results['threat'][j],
                'insult': batch_results['insult'][j],
                'identity_attack': batch_results['identity_attack'][j]
            }
            if all(score <= 0.1 for score in single_result.values()):
                is_safe = True
            else:
                is_safe = False
            
            if is_safe:
                safe_indices.append(idx)
    
    # Filter the original DataFrame
    filtered_df = df.iloc[safe_indices].copy()
    
    # Reset index to make it sequential
    filtered_df = filtered_df.reset_index(drop=True)
    
    # Save the filtered DataFrame
    filtered_df.to_json(output_file, orient='records', lines=True)
    
    print(f"Original dialog count: {len(df)}")
    print(f"Filtered dialog count: {len(filtered_df)}")
    print(f"Removed {len(df) - len(filtered_df)} toxic dialogs")
    
    return filtered_df, all_dialogs

if __name__ == "__main__":
    input_paths = ["/u/spa-d4/grad/mfe261/Projects/MobileConvRec/data/whole_dataset/amazon_beauty/serialised.jsonl", 
                   "/u/spa-d4/grad/mfe261/Projects/MobileConvRec/data/whole_dataset/amazon_electronics/serialised.jsonl",
                   "/u/spa-d4/grad/mfe261/Projects/MobileConvRec/data/whole_dataset/amazon_fashion/serialised.jsonl",
                   "/u/spa-d4/grad/mfe261/Projects/MobileConvRec/data/whole_dataset/amazon_garden/serialised.jsonl",
                   "/u/spa-d4/grad/mfe261/Projects/MobileConvRec/data/whole_dataset/amazon_grocery/serialised.jsonl",
                   "/u/spa-d4/grad/mfe261/Projects/MobileConvRec/data/whole_dataset/goodreads/serialised.jsonl",
                   "/u/spa-d4/grad/mfe261/Projects/MobileConvRec/data/whole_dataset/tripadvisor/serialised.jsonl",
                   "/u/spa-d4/grad/mfe261/Projects/MobileConvRec/data/whole_dataset/yelp/serialised.jsonl",
                   "/u/spa-d4/grad/mfe261/Projects/MobileConvRec/data/whole_dataset/mobilerec/serialised.jsonl"]
    output_paths = ["/u/spa-d4/grad/mfe261/Projects/MobileConvRec/data/whole_dataset/amazon_beauty/serialised_filtered.jsonl",
                    "/u/spa-d4/grad/mfe261/Projects/MobileConvRec/data/whole_dataset/amazon_electronics/serialised_filtered.jsonl",
                    "/u/spa-d4/grad/mfe261/Projects/MobileConvRec/data/whole_dataset/amazon_fashion/serialised_filtered.jsonl",
                    "/u/spa-d4/grad/mfe261/Projects/MobileConvRec/data/whole_dataset/amazon_garden/serialised_filtered.jsonl",
                    "/u/spa-d4/grad/mfe261/Projects/MobileConvRec/data/whole_dataset/amazon_grocery/serialised_filtered.jsonl",
                    "/u/spa-d4/grad/mfe261/Projects/MobileConvRec/data/whole_dataset/goodreads/serialised_filtered.jsonl",
                    "/u/spa-d4/grad/mfe261/Projects/MobileConvRec/data/whole_dataset/tripadvisor/serialised_filtered.jsonl",
                    "/u/spa-d4/grad/mfe261/Projects/MobileConvRec/data/whole_dataset/yelp/serialised_filtered.jsonl",
                    "/u/spa-d4/grad/mfe261/Projects/MobileConvRec/data/whole_dataset/mobilerec/serialised_filtered.jsonl"]
    
    for input_path, output_path in zip(input_paths, output_paths):
    
        filtered_df, all_dialogs = filter_toxic_dialogs(input_path, output_path)
        
        stats, freq_dist = analyze_token_lengths(all_dialogs)

        # Print detailed statistics
        print("Token Distribution Statistics:")
        print(f"Mean tokens per string: {stats['mean']:.2f}")
        print(f"Median tokens per string: {stats['median']:.2f}")
        print(f"Standard deviation: {stats['std']:.2f}")
        print(f"Range: {stats['min']} to {stats['max']} tokens")
        print(f"Total strings analyzed: {stats['total_strings']}")
        print(f"Total tokens across all strings: {stats['total_tokens']}")
        print(f"Average tokens per string: {stats['avg_tokens_per_string']:.2f}")
        print(f"More than 512 tokens: {stats['more_512']}")

        # Plot the distribution
        plot_token_distribution(freq_dist)
```

# Fluency Filtering

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import pandas as pd
from tqdm import tqdm
import torch
from collections import Counter
import matplotlib.pyplot as plt
import numpy as np
```

```python
def analyze_token_lengths(strings, tokenizer):
    """
    Analyze the distribution of token (word) lengths in a list of strings.
    Uses simple whitespace splitting for tokenization.
    
    Parameters:
    strings (list): List of strings to analyze
    
    Returns:
    tuple: (basic_stats, length_frequencies, total_tokens)
    """
    # Calculate token lengths for each string
    lengths = []
    total_tokens = 0
    for text in tqdm(strings):
        # Split on whitespace and filter out empty strings
        tokens = tokenizer.encode(text)
        num_tokens = len(tokens)
        lengths.append(num_tokens)
        total_tokens += num_tokens
    
    # Basic statistics
    stats = {
        'mean': np.mean(lengths),
        'median': np.median(lengths),
        'std': np.std(lengths),
        'min': min(lengths),
        'max': max(lengths),
        'total_strings': len(strings),
        'total_tokens': total_tokens,
        'avg_tokens_per_string': total_tokens / len(strings),
        "more_512": len([i for i in lengths if i > 512])
    }
    
    # Calculate frequency distribution
    freq_dist = Counter(lengths)
    
    # Convert to sorted list of tuples
    freq_dist = sorted(freq_dist.items())
    
    return stats, freq_dist

def plot_token_distribution(freq_dist):
    """
    Create a histogram of token counts.
    
    Parameters:
    freq_dist (list): List of (token_count, frequency) tuples
    """
    token_counts, freqs = zip(*freq_dist)
    
    plt.figure(figsize=(12, 6))
    plt.bar(token_counts, freqs)
    plt.xlabel('Number of Tokens (Words)')
    plt.ylabel('Frequency')
    plt.title('Distribution of Token Counts per String')
    plt.grid(True, alpha=0.3)
    
    # Add a rolling average line if there are enough data points
    if len(token_counts) > 5:
        window_size = 5
        rolling_avg = np.convolve(freqs, np.ones(window_size)/window_size, mode='valid')
        plt.plot(token_counts[window_size-1:], rolling_avg, 'r-', label='Rolling Average')
        plt.legend()
    
    plt.show()

def filter_non_fluent_dialogs(input_file, output_file, model, tokenizer, batch_size=64, device='cuda'):
    # Load the original DataFrame
    df = pd.read_json(input_file, lines=True)

    # Concatenate turns into a single string per dialog
    all_dialogs = []
    dialog_indices = []
    for idx, turns in enumerate(tqdm(df["turns"], desc="Processing dialogs")):
        concatenated_turns = []
        for entry in turns:
            if "COMPUTER" in entry:
                concatenated_turns.append(f"COMPUTER: {entry['COMPUTER']}")
            if "HUMAN" in entry:
                concatenated_turns.append(f"HUMAN: {entry['HUMAN']}")
        
        dialog = " ".join(concatenated_turns)
        all_dialogs.append(dialog)
        dialog_indices.append(idx)

    safe_indices = []
    model.to(device)
    
    for i in tqdm(range(0, len(all_dialogs), batch_size), desc="Checking fluency"):
        batch = all_dialogs[i:i + batch_size]
        batch_indices = dialog_indices[i:i + batch_size]
        
        with torch.no_grad():
            inputs = tokenizer(batch, return_tensors='pt', padding=True, truncation=True).to(device)
            outputs = model(**inputs)
            predictions = torch.softmax(outputs.logits, dim=1)
        
        for j, idx in enumerate(batch_indices):
            if predictions[j][0].item() > 0.6:
                safe_indices.append(idx)
            else:
                print(f"Dialog {idx} is non-fluent: {all_dialogs[idx]}, fluency: {predictions[j][0].item()}")
    
    filtered_df = df.iloc[safe_indices].reset_index(drop=True)
    filtered_df.to_json(output_file, orient='records', lines=True)
    
    print(f"Original dialog count: {len(df)}")
    print(f"Filtered dialog count: {len(filtered_df)}")
    print(f"Removed {len(df) - len(filtered_df)} non-fluent dialogs")
    
    return filtered_df, all_dialogs

if __name__ == "__main__":
    input_paths = ["/u/spa-d4/grad/mfe261/Projects/MobileConvRec/data/whole_dataset/amazon_beauty/serialised_filtered.jsonl", 
                   "/u/spa-d4/grad/mfe261/Projects/MobileConvRec/data/whole_dataset/amazon_electronics/serialised_filtered.jsonl",
                   "/u/spa-d4/grad/mfe261/Projects/MobileConvRec/data/whole_dataset/amazon_fashion/serialised_filtered.jsonl",
                   "/u/spa-d4/grad/mfe261/Projects/MobileConvRec/data/whole_dataset/amazon_garden/serialised_filtered.jsonl",
                   "/u/spa-d4/grad/mfe261/Projects/MobileConvRec/data/whole_dataset/amazon_grocery/serialised_filtered.jsonl",
                   "/u/spa-d4/grad/mfe261/Projects/MobileConvRec/data/whole_dataset/goodreads/serialised_filtered.jsonl",
                   "/u/spa-d4/grad/mfe261/Projects/MobileConvRec/data/whole_dataset/tripadvisor/serialised_filtered.jsonl",
                   "/u/spa-d4/grad/mfe261/Projects/MobileConvRec/data/whole_dataset/yelp/serialised_filtered.jsonl",
                   "/u/spa-d4/grad/mfe261/Projects/MobileConvRec/data/whole_dataset/mobilerec/serialised_filtered.jsonl"]
    output_paths = ["/u/spa-d4/grad/mfe261/Projects/MobileConvRec/data/whole_dataset/amazon_beauty/serialised_filtered_v2.jsonl",
                    "/u/spa-d4/grad/mfe261/Projects/MobileConvRec/data/whole_dataset/amazon_electronics/serialised_filtered_v2.jsonl",
                    "/u/spa-d4/grad/mfe261/Projects/MobileConvRec/data/whole_dataset/amazon_fashion/serialised_filtered_v2.jsonl",
                    "/u/spa-d4/grad/mfe261/Projects/MobileConvRec/data/whole_dataset/amazon_garden/serialised_filtered_v2.jsonl",
                    "/u/spa-d4/grad/mfe261/Projects/MobileConvRec/data/whole_dataset/amazon_grocery/serialised_filtered_v2.jsonl",
                    "/u/spa-d4/grad/mfe261/Projects/MobileConvRec/data/whole_dataset/goodreads/serialised_filtered_v2.jsonl",
                    "/u/spa-d4/grad/mfe261/Projects/MobileConvRec/data/whole_dataset/tripadvisor/serialised_filtered_v2.jsonl",
                    "/u/spa-d4/grad/mfe261/Projects/MobileConvRec/data/whole_dataset/yelp/serialised_filtered_v2.jsonl",
                    "/u/spa-d4/grad/mfe261/Projects/MobileConvRec/data/whole_dataset/mobilerec/serialised_filtered_v2.jsonl"]
    
    # Load model directly
    tokenizer = AutoTokenizer.from_pretrained("cointegrated/roberta-large-cola-krishna2020")
    model = AutoModelForSequenceClassification.from_pretrained("cointegrated/roberta-large-cola-krishna2020")
    
    for input_path, output_path in zip(input_paths, output_paths):
    
        filtered_df, all_dialogs = filter_non_fluent_dialogs(input_path, output_path, model, tokenizer)
        
        stats, freq_dist = analyze_token_lengths(all_dialogs, tokenizer)

        # Print detailed statistics
        print("Token Distribution Statistics:")
        print(f"Mean tokens per string: {stats['mean']:.2f}")
        print(f"Median tokens per string: {stats['median']:.2f}")
        print(f"Standard deviation: {stats['std']:.2f}")
        print(f"Range: {stats['min']} to {stats['max']} tokens")
        print(f"Total strings analyzed: {stats['total_strings']}")
        print(f"Total tokens across all strings: {stats['total_tokens']}")
        print(f"Average tokens per string: {stats['avg_tokens_per_string']:.2f}")
        print(f"More than 512 tokens: {stats['more_512']}")

        # Plot the distribution
        plot_token_distribution(freq_dist)
```

# Splitting the data to Train, Test, and Validations

```python
import json
import datetime
from typing import List, Dict
from pathlib import Path
from collections import defaultdict
```

```python
def load_jsonl(file_path: str) -> List[Dict]:
    """Load data from a JSONL file."""
    data = []
    with open(file_path, 'r', encoding='utf-8') as file:
        for line in file:
            try:
                data.append(json.loads(line.strip()))
            except json.JSONDecodeError as e:
                print(f"Error parsing line: {e}")
                continue
    return data

def extract_date(app_data: Dict) -> datetime.datetime:
    """Extract the date from the recommended app data."""
    return datetime.datetime.strptime(app_data['recommended_place']['date'], '%Y-%m-%dT%H:%M:%S')

def group_by_date(data: List[Dict]) -> Dict[datetime.datetime, List[Dict]]:
    """Group data entries by their dates."""
    date_groups = defaultdict(list)
    for entry in data:
        date = extract_date(entry)
        date_groups[date].append(entry)
    return date_groups

def split_data(data: List[Dict], train_ratio=0.7, val_ratio=0.15, test_ratio=0.15) -> tuple:
    """
    Split data into train, validation, and test sets based on dates,
    ensuring no date overlap between sets and maintaining approximate ratios.
    """
    # Group data by dates
    date_groups = group_by_date(data)
    
    # Get total count and cumulative counts by date
    total_count = len(data)
    target_train_count = int(total_count * train_ratio)
    target_val_count = int(total_count * val_ratio)
    
    # Sort dates and initialize splits
    sorted_dates = sorted(date_groups.keys())
    train_data = []
    val_data = []
    test_data = []
    
    current_count = 0
    train_end_idx = None
    val_end_idx = None
    
    # Find split points based on counts
    for i, date in enumerate(sorted_dates):
        current_count += len(date_groups[date])
        if train_end_idx is None and current_count >= target_train_count:
            train_end_idx = i
        if val_end_idx is None and current_count >= (target_train_count + target_val_count):
            val_end_idx = i
            break
    
    # If split points weren't found, use default positions
    if train_end_idx is None:
        train_end_idx = len(sorted_dates) - 2
    if val_end_idx is None:
        val_end_idx = len(sorted_dates) - 1
    
    # Distribute data ensuring no date overlap
    for i, date in enumerate(sorted_dates):
        entries = date_groups[date]
        if i <= train_end_idx:
            train_data.extend(entries)
        elif i <= val_end_idx:
            val_data.extend(entries)
        else:
            test_data.extend(entries)
    
    return train_data, val_data, test_data

def save_splits(train_data: List[Dict], val_data: List[Dict], test_data: List[Dict], output_dir: str):
    """Save the split data to separate JSONL files."""
    output_path = Path(output_dir)
    output_path.mkdir(parents=True, exist_ok=True)
    
    # Save each split to a separate file
    for name, data in [('train', train_data), ('val', val_data), ('test', test_data)]:
        file_path = output_path / f'{name}.jsonl'
        with open(file_path, 'w', encoding='utf-8') as f:
            for item in data:
                f.write(json.dumps(item) + '\n')

def analyze_split(train_data, val_data, test_data):
    """Analyze the data split and return summary statistics."""
    def get_date_range(data):
        if not data:
            return None, None
        dates = [extract_date(d) for d in data]
        return min(dates), max(dates)

    def get_stats(data):
        if not data:
            return {'count': 0, 'date_range': (None, None), 'percentage': 0}
        
        date_range = get_date_range(data)
        return {
            'count': len(data),
            'date_range': date_range,
            'percentage': len(data) / total_count * 100,
            'unique_dates': len({extract_date(d) for d in data})
        }

    total_count = len(train_data) + len(val_data) + len(test_data)
    
    stats = {
        'train': get_stats(train_data),
        'validation': get_stats(val_data),
        'test': get_stats(test_data),
        'total': total_count
    }
    
    return stats
```

```python
input_file = "/u/spa-d4/grad/mfe261/Projects/MobileConvRec/dataset/yelp/serialised_filtered.jsonl"
output_dir = "/u/spa-d4/grad/mfe261/Projects/MobileConvRec/dataset/yelp/splits"

print(f"Loading data from {input_file}...")
data = load_jsonl(input_file)
print(f"Loaded {len(data)} records")
    
# Split data
print("\nSplitting data...")
train_data, val_data, test_data = split_data(data)
    
# Save splits
print(f"\nSaving splits to {output_dir}...")
save_splits(train_data, val_data, test_data, output_dir)
    
# Analyze and print results
stats = analyze_split(train_data, val_data, test_data)
    
print("\nData Split Analysis:")
print(f"Total records: {stats['total']}")
for split_name in ['train', 'validation', 'test']:
    split_stats = stats[split_name]
    print(f"\n{split_name.capitalize()} Set:")
    print(f"Count: {split_stats['count']} ({split_stats['percentage']:.1f}%)")
    print(f"Date Range: {split_stats['date_range'][0].date()} to {split_stats['date_range'][1].date()}")
```
